```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Graph 1**

```python
tables = pd.read_excel ('/CO2 Emissies (bewerkt).xlsx', skiprows=3)
countries_CO2=tables
countries_CO2
```

```python
countries_CO2.columns = ['country', '1990', '2005', '2017', '2021','2017(% of world)', '2017 vs 1990: change (%)', 'per land area', 'per capita', 'total inc_lucf', 'total ex_lucf']
countries_CO2
```

```python
countries_CO2 = countries_CO2.loc[3:]
countries_CO2 = countries_CO2.drop([67])
countries_CO2
```

```python
from prompt_toolkit.formatted_text.utils import to_plain_text
top_5_CO2 = countries_CO2.loc[:,['country','1990','2005','2017','2021']]
top_5_CO2
```

```python
top_5 = top_5_CO2.sort_values(by='2021',ascending=False).head(5)
top_5
```

```python
fig, ax =plt.subplots()
years = top_5.columns[1:]
for index, row in top_5.iterrows():
  plt.plot(years, row[1:], label=row[0])

plt.title('Top 5 CO2 producers')
plt.xlabel('Year')
plt.ylabel('Production')
plt.legend(loc='center left', bbox_to_anchor=(-1,0.5))
plt.show()
```

**Graph 2**

```python
table_2 = countries_CO2.loc[:,["country", "1990", "2005", "2017", "2021"]]
table_2 = table_2.loc[table_2["1990"] >= 5]
table_2
```

```python

pd.options.display.float_format = '{:.3f}'.format
#New colums with percentage of change per year based on 1990.
table_2['1990vs1990ch'] = table_2.loc[:,'1990'] / table_2.loc[:,'1990'] * 100
table_2['2005vs1990ch'] = table_2.loc[:,'2005'] / table_2.loc[:,'1990'] * 100
table_2['2017vs1990ch'] = table_2.loc[:,'2017'] / table_2.loc[:,'1990'] * 100
table_2['2021vs1990ch'] = table_2.loc[:,'2021'] / table_2.loc[:,'1990'] * 100
table_2 = table_2.loc[:,["country","1990vs1990ch", "2005vs1990ch", "2017vs1990ch", "2021vs1990ch"]] 
table_2
table_2_best = table_2.sort_values(by='2021vs1990ch', ascending=True).head(3) 
table_2_worst = table_2.sort_values(by='2021vs1990ch', ascending=False).head(3) 
table_2_best.columns = ["country", "1990", "2005", "2017", "2021"]
table_2_best
```

```python
fig, ax =plt.subplots()
years = table_2_best.columns[1:]
for index, row in table_2_best.iterrows():
  plt.plot(years, row[1:], label=row[0])

plt.title('Most percentage of decrease changes in the world')
plt.xlabel('Year')
plt.ylabel('CO2 change in percentage')
plt.legend(loc='center left', bbox_to_anchor=(-1,0.5))
plt.show()
     
```

```python
table_2_worst.columns = ["Country", "1990", "2005", "2017", "2021"]# Renaming for better use in graph.
table_2_worst
```

```python
fig, ax =plt.subplots()
years = table_2_worst.columns[1:]
for index, row in table_2_worst.iterrows():
  plt.plot(years, row[1:], label=row[0])

plt.title('Most percentage of increase changes in the world')
plt.xlabel('Year')
plt.ylabel('CO2 change in percentage')
plt.legend(loc='center left', bbox_to_anchor=(-1,0.5))
plt.show()
```
